# Instagram Media Data Cleaning

```python
import os
import pandas as pd
from dateutil import parser
from datetime import datetime, timedelta
```

```python
# collate data files
files = ["dnssingapore.txt","thewokesalaryman.txt", "seedlysg.txt", "mortgageconsultancy.txt"]

data = pd.read_csv('dnssingapore.txt', delimiter = "\t")
data = data.append(pd.read_csv('thewokesalaryman.txt', delimiter = "\t"))
data = data.append(pd.read_csv('seedlysg.txt',  delimiter = "\t"))
data = data.append(pd.read_csv('mortgageconsultancy.txt', delimiter = "\t"))

data.shape
```

```python
# format date to GMT +8
data['date']=data['date'].apply(lambda x: parser.parse(x)+timedelta(hours=8))
data.head()
```

```python
# format to YYYYMMDDThhmm
data['timestamp']=data['date'].apply(lambda x: x.strftime("%Y%m%dT%H%M"))
```

```python
data.to_csv(os.getcwd()+"/InstagramLikesComments.txt",sep='\t',encoding='utf-8')
```
