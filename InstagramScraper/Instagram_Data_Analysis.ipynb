# Getting tagging network

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
```

## dnssingapore

```python
dns_tagged_comments = dns_comments_tags[['comment','commenter']]
dns_tagged_comments
```

```python
dns_tagged_comments['comment'] = dns_tagged_comments['comment'].apply(lambda x : re.findall("@\w+",x))
dns_tagged_comments
```

```python
vectorizer = CountVectorizer(analyzer = (lambda x: x))
dns_tags = pd.DataFrame(vectorizer.fit_transform(dns_tagged_comments['comment']).toarray(), 
                          columns=vectorizer.get_feature_names())
dns_tags
```

```python
dns_tags['commenter'] = dns_tagged_comments['commenter'].tolist()
dns_tags
```

```python
dns_tags = dns_tags.set_index('commenter') \
                        .unstack() \
                        .reset_index()
dns_tags.head()
```

```python
dns_tags.columns = ['tagged_user', 'commenter', 'weight']
dns_tags = dns_tags.loc[dns_tags['weight'] > 0]
dns_tags['tagged_user'] = dns_tags['tagged_user'].apply(lambda x: x[1:])
dns_tags.sort_values('commenter')
```

## seedlysg

```python
seedlysg_tagged_comments = seedlysg_comments_tags[['comment','commenter']]
seedlysg_tagged_comments
```

```python
seedlysg_tagged_comments['comment'] = seedlysg_tagged_comments['comment'].apply(lambda x : re.findall("@\w+",x))
seedlysg_tagged_comments
```

```python
vectorizer = CountVectorizer(analyzer = (lambda x: x))
seedlysg_tags = pd.DataFrame(vectorizer.fit_transform(seedlysg_tagged_comments['comment']).toarray(), 
                          columns=vectorizer.get_feature_names())
seedlysg_tags
```

```python
seedlysg_tags['commenter'] = seedlysg_tagged_comments['commenter'].tolist()
seedlysg_tags
```

```python
seedlysg_tags = seedlysg_tags.set_index('commenter') \
                        .unstack() \
                        .reset_index()
seedlysg_tags.head()
```

```python
seedlysg_tags.columns = ['tagged_user', 'commenter', 'weight']
seedlysg_tags = seedlysg_tags.loc[seedlysg_tags['weight'] > 0]
seedlysg_tags['tagged_user'] = seedlysg_tags['tagged_user'].apply(lambda x: x[1:])
seedlysg_tags.sort_values('commenter')
```

```python
# sum weights
consolidated_seedlysg_tags = {}
for index, row in seedlysg_tags.iterrows():
    key = (row['commenter'],row['tagged_user'])

    if key in consolidated_seedlysg_tags:
        consolidated_seedlysg_tags[key] += 1
    else:
        consolidated_seedlysg_tags[key] = 1
consolidated_seedlysg_tags
```

```python
# Start writing a GML output file
output_file = open("seedlysg_tags.gml", "w")
output_file.write("graph\n")
output_file.write("[\n")
```

```python
# print all nodes
id_count = 1
username_ids = {}
for username in np.unique(seedlysg_tags[['tagged_user', 'commenter']].values):
    output_file.write("  node\n")
    output_file.write("  [\n")
    output_file.write("    id " + str(id_count) + "\n")
    output_file.write("    label \"" + username + "\"" + "\n")
    output_file.write("  ]\n")
    
    username_ids[username] = id_count
    id_count += 1

# print all edges
for keys in consolidated_seedlysg_tags.keys():
    output_file.write("  edge\n")
    output_file.write("  [\n")
    output_file.write("    source " + str(username_ids[keys[0]]) + "\n")
    output_file.write("    target " + str(username_ids[keys[1]]) + "\n")
    output_file.write("    value " + str(consolidated_seedlysg_tags[keys]) + "\n")
    output_file.write("  ]\n")

output_file.write("]\n")
output_file.close()
```
